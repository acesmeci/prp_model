### Sim Study 1 -- Quantifying Representational Overlap

```python
import numpy as np
from collections import defaultdict

# Compute Mean Hidden Activations per Task
def get_mean_hidden_activations(task_net, stim_inputs, task_inputs, task_ids, tau_net=1.0):
    """
    Returns a dictionary of task_id -> mean hidden activation vector.
    """
    hidden_by_task = defaultdict(list)

    for stim, task, task_id in zip(stim_inputs, task_inputs, task_ids):
        stim_t = torch.tensor(stim, dtype=torch.float32)
        task_t = torch.tensor(task, dtype=torch.float32)

        h = task_net.get_hidden_activation(stim_t, task_t, tau_net)
        hidden_by_task[int(task_id)].append(h)

    mean_hidden = {}
    for task_id, acts in hidden_by_task.items():
        mean_hidden[task_id] = np.mean(acts, axis=0)

    return mean_hidden

# Compute Task Similarity Matrix
def compute_task_similarity_matrix(mean_hidden):
    """
    Returns a matrix of Pearson r correlations between all tasks.
    """
    task_ids = sorted(mean_hidden.keys())
    activations = np.stack([mean_hidden[tid] for tid in task_ids])
    corr_matrix = np.corrcoef(activations)
    return task_ids, corr_matrix

# Look Up Similarity Between Two Tasks
def compare_task_pair_similarity(task_id_a, task_id_b, task_ids, corr_matrix):
    idx_a = task_ids.index(task_id_a)
    idx_b = task_ids.index(task_id_b)
    return corr_matrix[idx_a, idx_b]


```

```python
# Plot task similarity matrix

# Generate data for a few tasks
input_sgl, tasks_sgl, _, meta = generate_task_patterns(
    N_pathways=3, N_features=3, samples_per_task=10, relevant_tasks=[1, 2, 4, 5, 9], seed=42
)

task_ids = meta["tasks_idx"]  # Task ID for each sample
mean_hidden = get_mean_hidden_activations(net, input_sgl, tasks_sgl, task_ids)
task_ids_sorted, sim_matrix = compute_task_similarity_matrix(mean_hidden)

# Visualize or inspect
import seaborn as sns
import matplotlib.pyplot as plt

import os

# Ensure output/plots directory exists
os.makedirs("output/plots", exist_ok=True)

plt.figure(figsize=(6,5))
sns.heatmap(sim_matrix, xticklabels=task_ids_sorted, yticklabels=task_ids_sorted, annot=True, cmap='viridis')
plt.title("Hidden Layer Task Similarity (Pearson r)")
plt.savefig("output/plots/task_similarity_matrix.png")
plt.show()

# Optional: print specific pair similarity
print("Similarity between Task 1 and Task 4:", compare_task_pair_similarity(1, 4, task_ids_sorted, sim_matrix))
```

#### Multitask trained task similarity matrix

```python

input_multi, tasks_multi, _, meta = generate_multitask_patterns(
    N_pathways=3,
    N_features=3,
    samples_per_task=10,
    relevant_tasks=[1, 2, 4, 5, 9]
)

task_ids_multi = meta["taskIdx"]  # Gives ID of each task combination

# Split multi-task vectors into two individual tasks
def extract_individual_tasks(task_vector):
    return np.where(task_vector == 1)[0]  # e.g., [1, 4] if task_vector = [0,1,0,0,1,0,...]

# Build per-task dataset from multitask inputs
def split_multitask_data(input_multi, tasks_multi):
    stim_list = []
    task_list = []
    task_ids = []

    for stim, task_vec in zip(input_multi, tasks_multi):
        task_ids_this = extract_individual_tasks(task_vec)
        for tid in task_ids_this:
            task_input = np.zeros_like(task_vec)
            task_input[tid] = 1
            stim_list.append(stim)
            task_list.append(task_input)
            task_ids.append(tid + 1)  # +1 to match MATLAB-style IDs

    return np.array(stim_list), np.array(task_list), np.array(task_ids)

# RUn hidden similarity analysis on multitask data
stim_split, task_split, task_ids_split = split_multitask_data(input_multi, tasks_multi)

mean_hidden_multi = get_mean_hidden_activations(net, stim_split, task_split, task_ids_split)
task_ids_sorted, sim_matrix_multi = compute_task_similarity_matrix(mean_hidden_multi)

# Visualize
plt.figure(figsize=(6, 5))
sns.heatmap(sim_matrix_multi, xticklabels=task_ids_sorted, yticklabels=task_ids_sorted, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Task Similarity Matrix After Multitask Training (Hidden Layer)")
plt.tight_layout()
plt.savefig("output/plots/task_similarity_matrix_multitask.png")
plt.show()

```

```python
def simple_trial_generator():
    input_sgl, tasks_sgl, _, _ = generate_task_patterns(
        N_pathways=3, N_features=3, samples_per_task=1,
        relevant_tasks=[1, 4], seed=None  # high interference pair
    )
    return input_sgl[0], input_sgl[1], tasks_sgl[0], tasks_sgl[1]



# Generate high-interference multitask trial
input_a, input_b, task_a, task_b = simple_trial_generator()
soa = 0  # simultaneous onset

input_series = []
task_series = []

for t in range(50):
    stim_t = np.zeros_like(input_a)
    task_t = np.zeros_like(task_a)
    
    if t >= 0:      # Task A starts at t=0
        stim_t += input_a
        task_t += task_a
    if t >= soa:    # Task B also starts at t=0
        stim_t += input_b
        task_t += task_b

    input_series.append(stim_t)
    task_series.append(task_t)

# Use stronger tau_task
tau_task = 1.0
net_input_series = net.get_output_net_input_series(input_series, task_series, tau_task=tau_task)

# Sum total net input across all output units for each time step
total_input = [np.sum(v) for v in net_input_series]

# Plot
plt.plot(total_input)
plt.title("Total Net Input to Output Layer (Task 1 + Task 4)")
plt.xlabel("Time Step")
plt.ylabel("Sum of Net Output Input")
plt.grid(True)
plt.show()

```

```python
def visualize_task_b_activations(task_net, trial_generator, soa_values, max_timesteps=50, tau_net=1.0, tau_task=1.0):
    # Use one fixed trial to eliminate variability
    input_a, input_b, task_a, task_b = trial_generator()

    # Get Task B output indices
    N_pathways = 3
    N_features = 3
    task_matrix_b = task_b.reshape(N_pathways, N_pathways).T
    in_dim_b, out_dim_b = np.argwhere(task_matrix_b == 1)[0]
    relevant_outputs_b = list(range(out_dim_b * N_features, (out_dim_b + 1) * N_features))

    plt.figure(figsize=(10, 6))
    for soa in soa_values:
        # Generate input series with current SOA
        input_dim = input_a.shape[0]
        task_dim = task_a.shape[0]
        input_series = []
        task_series = []

        for t in range(max_timesteps):
            stim_t = np.zeros(input_dim)
            task_t = np.zeros(task_dim)

            if t >= 0:
                stim_t += input_a
                task_t += task_a

            if t >= soa:
                stim_t += input_b
                task_t += task_b

            input_series.append(stim_t)
            task_series.append(task_t)

        # Forward pass
        output_series = task_net.integrate(input_series, task_series,
                                           tau_net=tau_net, tau_task=tau_task)

        # Extract relevant Task B output over time
        task_b_activations = [output[relevant_outputs_b].mean() for output in output_series]
        plt.plot(task_b_activations, label=f"SOA = {soa}")

    plt.xlabel("Time Step")
    plt.ylabel("Mean Activation (Task B Output Units)")
    plt.title("Task B Output Activation Over Time")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
